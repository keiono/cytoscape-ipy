## Analyze Graphs Using NetworkX and Cytoscape.js

```python
from IPython.html.widgets import interact
from IPython.html import widgets
import cytoscape.viewer as cy, json
import networkx as nx
from igraph import *
```

```python
yeast_network = json.load(open('yeast2.cyjs'))
vs_collection = json.load(open('style.json'))

styles = {}
for style in vs_collection:
    style_settings = style['style']
    title = style['title']
    styles[title] = style_settings
```

```python
def render_graph(Style, Layout):
    cy.render(yeast_network, Style, Layout)

interact(render_graph, Style = styles, Layout = cy.get_layouts())
```

```python
# Generate Graph with NetworkX
netx = nx.barabasi_albert_graph(100,3)
num_nodes = ba_graph1.number_of_nodes()
num_edges = ba_graph1.number_of_edges()
print('Nodes = ' + str(num_nodes))
print('Edges = ' + str(num_edges))

cyjs1 = cy.from_networkx(netx)

def render_networkx_graph(Style, Layout):
    cy.render(cyjs1, Style, Layout)

interact(render_networkx_graph, Style = styles, Layout = cy.get_layouts())
```

```python
# Generate with igraph
generated1 = Graph.Barabasi(300, 3)
layout = generated1.layout("kk")

cyjs2 = cy.from_igraph(generated1, layout, 200)


# Override the existing style
new_style = styles['default']
new_style.append({
    'selector':'node', 
    'css':{
        'width':20,
        'height':20
    }
})

new_style.append({
    'selector':'edge', 
    'css':{
        'width':1
    }
})
cy.render(cyjs2, new_style, 'preset')
```
